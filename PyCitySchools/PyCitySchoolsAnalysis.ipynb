# PyCity Schools Analysis

- The PyCity School District consists of 15 high schools (grades 9-12) with a total student body of 39,170 (average of 2,611.33 students per school).  Eight of the schools are Charter schools.
- The District-wide passing rates for reading are significantly higher than those for math with 85.80% of district-wide students passing reading (or 33,608 students) and 74.98% passing math (or 29,370 students).
- The District-wide average budget per student is $629.29, but this varies considerably by school. 

  - The lowest school budget per student is $578 
  
  - The highest school budget per student is $655.
  
  - That is a $77 per student difference between the highest and lowest funded school.
  
- All of the Charter schools within the PyCity District are lower in student population (ranging between 427 to 2,283) than the District schools (ranging between 2,739 to 4,976).
- All but one of the Charter schools are lower in per student budget (ranging between 578 to 625) than the District schools (ranging between 628 to 655). 

  - Johnson HS, a Charter school, has a per student budget of $638.
  
- The overall passing percentages between the District and Charter schools is alarming:

  - District oveall passing percentages range between 52.99% (3,999 students, 637 per student) to 54.64% (4,976 students, $628 per student).
  
  - Charter schools overall passing percentages range between 89.23% (427 students, 581 per student) to 91.33% (1,858 students, $582 per student).
  
- District students passing percentages for reading and math vary greatly (~ 80% vs ~ 66%) while Charter students passing percentages for reading and math vary little (~ 90% vs ~ 96%).

In Conclusion:
- High student achievement seems to be more correlated with lower student bodies than with higher budget.
- Charter schools consistently perform better than District schools, and in all measured areas.
- What this data does not tell us is teacher to student ratio, attendance rates, household incomes, and other criteria that could impact student achievement.

```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# Load files & store filepaths in a variable
school_file = Path("Resources/schools_complete.csv")
students_file = Path("Resources/students_complete.csv")

# Read and display CSVs with Pandas
schools_data = pd.read_csv(school_file)
students_data = pd.read_csv(students_file)
schools_data.head()
```

```python
students_data.head()
```

```python
# Merge schools & students together based on the shared school name
merged_data_df = pd.merge(schools_data, students_data, how="left", on=["school_name"])
merged_data_df
```

# District Summary¶

```python
school_count = merged_data_df['school_name'].nunique()
school_count
```

```python
student_count = merged_data_df['student_name'].count()
student_count
```

```python
school_budget = merged_data_df.groupby(['school_name', 'budget']).sum()
school_budget = school_budget.reset_index()
total_budget = school_budget['budget'].sum()
total_budget
```

```python
avg_math_score = merged_data_df["math_score"].mean()
avg_math_score
```

```python
avg_read_score = merged_data_df["reading_score"].mean()
avg_read_score
```

```python
passing_math = merged_data_df[(merged_data_df["math_score"] >= 70)].count()["student_name"]
passing_math
```

```python
passing_math_percent = passing_math / float(student_count) * 100
passing_math_percent
```

```python
passing_reading = merged_data_df[(merged_data_df["reading_score"] >= 70)].count()["student_name"]
passing_reading
```

```python
passing_read_percent = passing_reading / float(student_count) * 100
passing_read_percent
```

```python
passing_math_reading_count = merged_data_df[
    (merged_data_df["math_score"] >= 70) & (merged_data_df["reading_score"] >= 70)
].count()["student_name"]
overall_passing = passing_math_reading_count /  float(student_count) * 100
overall_passing
```

```python
# High-level snapshot of district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                    "Total Students": [student_count],
                                    "Total Budget": [total_budget],
                                    "Average Math Score": [avg_math_score],
                                    "Average Reading Score": [avg_read_score],
                                    "% Passing Math": [passing_math_percent],
                                    "% Passing Reading": [passing_read_percent],
                                    "% Overall Passing": [overall_passing]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace('$', '').astype(float)

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels)

school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

# Assemble into DataFrame
spending_summary_data = {'Average Math Score': spending_math_scores,
                         'Average Reading Score': spending_reading_scores,
                         '% Passing Math': spending_passing_math,
                         '% Passing Reading': spending_passing_reading,
                         '% Overall Passing': overall_passing_spending}

spending_summary = pd.DataFrame(spending_summary_data)

# Display results
spending_summary
```

# Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

school_spending_df["School Size"] = pd.cut(school_spending_df["Total Students"], size_bins, labels=labels)

school_spending_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_spending_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_spending_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_spending_df.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_spending_df.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_spending_df.groupby(["School Size"])["% Overall Passing"].mean()

# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`

size_summary_data = {'Average Math Score': size_math_scores,
                     'Average Reading Score': size_reading_scores,
                     '% Passing Math': size_passing_math,
                     '% Passing Reading': size_passing_reading,
                     '% Overall Passing': size_overall_passing}

size_summary = pd.DataFrame(size_summary_data)

# Display results
size_summary
```
